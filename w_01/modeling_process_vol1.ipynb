## 特徴量の作成
都心から離れるほど安いなど、位置と住宅価格は関係があると思われる為、緯度、経度と価格を深掘りする。  
対数化された価格に対して

    ・    〜 0% 	11.225243
    ・0 % 〜 25% 	12.682152
    ・25% 〜 50% 	13.017003
    ・50% 〜 75% 	13.377006
    ・75% 〜 100% 15.856731
    ・100%(外れ値） 〜 
    
    
で分類し、price_rank_cdとして列を追加する。

```python
def get_price_rank_cd(arg):

    '''
        価格を4等分する。
        arg pandas.core.series.Series
    '''
    price_range_cd = None
    price = arg["price_log"]
    long = arg["long"]
    lat = arg["lat"]
    
    if price and price < 12.682152:
        price_range_cd = 1
    elif 12.682152 <= price and price < 13.017003:
        price_range_cd = 2
    elif 13.017003 <= price and price < 13.377006:
        price_range_cd = 3
    elif 13.377006 <= price:
        price_range_cd = 4
    return price_range_cd

# price_rank_cdを追加（axis=1で各行に関数を適用）
house_dt_temp["price_rank_cd"] = house_dt_temp.apply(get_price_rank_cd, axis=1)

# price_rank_cdを確認（一番最後の列）
house_dt_temp.head()
```

再度、緯度、経度をprice_rank_cdを加えてプロット

```python
sns.lmplot( x="long", y="lat", hue="price_rank_cd", data=house_dt_temp, fit_reg=False, aspect=2)
```

price_rank_cd = 4 が固まっている地域がわかる。  
この地域の中心を求め、この中心から離れるほど、価格が下がると予想。

price_rank_cd = 4 と price_rank_cd = 1 のみを取り出し4分割してプロットしてみる。

```python
# price_rank_cd = 4 のデータ
house_dt_high = house_dt_temp[house_dt_temp["price_rank_cd"] == 4]

# 21613 / 4 =  5403 
house_dt_high.shape
```

```python
# price_rank_cd = 1 のデータ 
house_dt_low = house_dt_temp[house_dt_temp["price_rank_cd"] == 1]

# 21613 / 4 =  5403 
house_dt_high.shape
```

```python
# 期待値は 5403 だったので、他のレコード数も確認
print("### price_rank_cd = 2: num = {}".format(house_dt_temp[house_dt_temp["price_rank_cd"] == 2].shape))
print("### price_rank_cd = 3: num = {}".format(house_dt_temp[house_dt_temp["price_rank_cd"] == 3].shape))
```

それぞれ、4分割して再度プロット

```python
high_group_names = ["high_1", "high_2", "high_3", "high_4"]

# 4分割してprice_rank_cd列に再代入
house_dt_high["price_rank_cd"] = pd.cut(house_dt_high["price_log"], 4, precision=2, labels=high_group_names)

sns.lmplot( x="long", y="lat", hue="price_rank_cd", data=house_dt_high, fit_reg=False, aspect=2)
```

住宅価格が低い方は4分割しても偏りがみられなかったので、8分割してみる。    
下記のwarningが表示されたので

> A value is trying to be set on a copy of a slice from a DataFrame.
> Try using .loc[row_indexer,col_indexer] = value instead

_`loc`_ を使用して代入   

```python
low_group_names = [
    "low_1", "low_2", "low_3", "low_4", "low_5", "low_6", "low_7", "low_8"
]

# 8分割してprice_rank_cd列に再代入
house_dt_low.loc[:, "price_rank_cd"] = pd.cut(
    house_dt_low.loc[:, "price_log"], 8, precision=2, labels=low_group_names)

sns.lmplot(
    x="long",
    y="lat",
    hue="price_rank_cd",
    data=house_dt_low,
    fit_reg=False,
    aspect=2)
```

__緯度：47.6、経度：-122.2近辺が高そうなので、その近辺で最高額の住宅の緯度、経度を調べる__

緯度:47.6298  
経度：-122.323  

から離れるほど安いと仮定。

また、相関係数から見ても、緯度の方が価格に対して敏感なので、経度に関しては、緯度、経度の、最大 - 最小値の 比から1/2を乗じたペナルティーを与える。

```python
# 緯度の最大 - 最小
47.777600 - 47.155900
```

```python
# 経度の最大 - 最小
-121.315000 + 122.519000
```

```python
def distance_from_center(arg):

    x = (arg["long"] + 122.323) / 2
    y = arg["lat"] - 47.6298

    return np.sqrt(x**2 + y**2)

# distance列に値を割り当て（axis=1で各行に関数を適用）
house_dt["distance"] = house_dt.apply(distance_from_center, axis=1)
```

新しい特徴量では、強い負の相関があり、距離が遠くなるほど、価格が下がる事が分かる。

```python
#相関係数を算出
house_dt.corr().loc["price_log", "distance"]
```

```python
# price_logを縦軸に各項目をプロット
sns.lmplot(
    x="distance",
    y="price_log",
    hue="grade",
    data=house_dt,
    fit_reg=False,
    aspect=2)
```

sqrt_livingの様にきれいな分布ができていないが、負の相関の傾向は見受けられる。

### 変数選択

変数増加法で変数選択をしてみる  
変数増加法：一番説明力の高い特徴量(feature_importances_)から始めて、説明力の高い順番でRMSEを計算し、少なくなり続ける限り特徴量として採用していく。

```python
# 特徴量の重要度（feature_importances_）をDataFrame化
feature_importances = pd.DataFrame(rf.feature_importances_, columns=['feature_importances_'], index=feature_cols)

# 重要度でソート
feature_importances = feature_importances.sort_values(by="feature_importances_", ascending=False)
feature_importances
```

```python
# 変数増加法が行いやすいように重要度順に列名を入れ替え
reorder_house_dt = house_dt[feature_cols].reindex(columns = feature_importances.index.values)
reorder_house_dt[:3]
```

```python
# 標準化
reorder_X  = scaler.fit_transform(reorder_house_dt)

# sklearnに投入するため、NumPyの配列化
reorder_X = np.array(reorder_X)
```

```python
X_train_, X_test_, y_train_, y_test_ = train_test_split(
    reorder_X, y, test_size=0.3, random_state=1234)
```

```python
# デバッグの設定（使用したい箇所の直前に挿入
# from IPython.core.debugger import Pdb; Pdb().set_trace()


def get_gfs_feature_indices(X, y, feature_names, clf):
    """
    変数増加法による変数選択
    RMSEが小さくなる限り、特徴量として採用する。
    
    Args：
        X (numpy.ndarray): 列名を重要度順並べ替えた 特徴量データ
        y (numpy.ndarray): 教師データ
        feature_names(numpy.ndarray): 重要度順に並べた特徴量名
        clf : sklearnの学習モデル
    
    Returns:
        numpy.ndarray：採用した特徴量名の一覧
    """

    # 学習データを70%(X_train, y_train)、テストデータを30%(X_test, y_test)に分割にする。
    X_train_, X_test_, y_train_, y_test_ = train_test_split(
        X, y, test_size=0.3, random_state=1234)

    # 評価（MSE）を無限大で初期化
    last_rmse = np.inf

    #選ばれた特徴量の名前一覧
    chosen_features = []

    # 重要度順にランダムフォレストを試す
    for i, col_name in enumerate(feature_names):

        chosen_features.append(col_name)
        size = len(chosen_features)

        # chosen_feature にある特徴量のみ使用して学習 (全行の len(chosen_features)目まで)
        clf.fit(X_train_[:, :size], y_train_)

        # 予測
        y_pred = clf.predict(X_test_[:, :size])

        # RMSE（二乗平均平方根誤差 ）
        rmse = np.sqrt(mean_squared_error(y_test_, y_pred))

        print('Newly Added Feature: {},\tRMSE Score: {}'.format(
            col_name, np.sqrt(rmse)))

        # 特徴量を増やす事で、RMSEが減らなくなったら終了
        if rmse >= last_rmse:
            break

        #直近のRMSEを更新
        last_rmse = rmse

    return chosen_features
```

```python
# 各特徴量のRMSE（二乗平均平方根誤差）を出してみる
selected_feature_names = get_gfs_feature_indices(
    X=reorder_X,
    y=y,
    feature_names=feature_importances.index.values,
    clf=RandomForestRegressor(n_estimators=10, random_state=1234))
```

### 変数選択後の精度を再確認

モデル選択時と同じデータを使用して再確認

```python
# 列の並び順が変わっているので、データを分割
X_train_, X_test_, y_train_, y_test_ = train_test_split(reorder_X, y, test_size=0.3, random_state=1234)
X_train1_, X_train2_, y_train1_, y_train2_ = train_test_split(X_train_, y_train_, test_size=0.3, random_state=1234)

size = len(selected_feature_names)

rf = RandomForestRegressor(n_estimators=10, random_state=1234)

rf.fit(X_train1_[:, :size], y_train1)  # 学習
y_pred2 = rf.predict(X_train2[:, :size])  # 予測

# MSE（平均２乗誤差）を計算
rf_mse = mean_squared_error(y_train2, y_pred2)

# 二乗平均平方根誤差
print('RandomForest RMSE: ', np.sqrt(rf_mse))

print("\n[X_train1] Accuracy : {:.2f}".format(rf.score(X_train1_[:, :size], y_train1_)))
print("[X_train2] Accuracy : {:.2f}".format(rf.score(X_train2_[:, :size], y_train2_)))
```
